```python
# Guarda el original por seguridad
df_amazon_product["category_raw"] = df_amazon_product["category"].astype(str)

# Split por jerarqu√≠a
parts = df_amazon_product["category_raw"].str.split("|")

# Nivel superior (top) y hoja (leaf)
df_amazon_product["category_top"]  = parts.str[0]
df_amazon_product["category_leaf"] = parts.str[-1]

print("Ejemplo:")
print(df_amazon_product[["category_raw","category_top","category_leaf"]].head(5))
```

```python
import re

def clean_token(s: str) -> str:
    s = str(s)
    s = s.replace("&", " & ")  # arregla cosas como "Mobiles&Accessories"
    s = s.replace(",", ", ")   # comas con espacio
    s = re.sub(r"(?<=[a-z])(?=[A-Z])", " ", s)  # ComputerAccessories ‚Üí Computer Accessories
    s = re.sub(r"\s+", " ", s).strip()
    return s.title()  # pone cada palabra con may√∫scula

df_amazon_product["category_top_clean"]  = df_amazon_product["category_top"].map(clean_token)
df_amazon_product["category_leaf_clean"] = df_amazon_product["category_leaf"].map(clean_token)
```

```python
top_map = {
    "Computers & Accessories": "Computers & Accessories",
    "Electronics": "Electronics",
    "Home & Kitchen": "Home & Kitchen",
    "Office Products": "Office Products",
    "Musical Instruments": "Musical Instruments",
    "Home Improvement": "Home Improvement",
    "Toys & Games": "Toys & Games",
    "Health & Personal Care": "Health & Personal Care",
    "Car & Motorbike": "Car & Motorbike",
}

df_amazon_product["category_top_clean"] = df_amazon_product["category_top_clean"].apply(
    lambda x: top_map.get(x, x)  # deja igual si no est√° mapeado
)
```

## Introducci√≥n

En el competitivo entorno del comercio electr√≥nico, las estrategias de precios juegan un rol clave en la percepci√≥n y decisi√≥n de compra de los consumidores. Una t√°ctica com√∫n es aplicar descuentos visibles en los productos, lo cual puede influir tanto en el volumen de ventas como en las calificaciones que reciben los productos.

Este proyecto tiene como objetivo responder a una pregunta de negocio concreta:

 **¬øCu√°nto mayor es el descuento, mejor es el rating del producto en Amazon?**

A partir de esta pregunta, se busca analizar si existe una relaci√≥n significativa entre el porcentaje de descuento aplicado a un producto y la calificaci√≥n promedio (rating) otorgada por los usuarios. La respuesta puede ayudar a mejorar las estrategias comerciales, equilibrando descuentos agresivos con la reputaci√≥n del producto y el comportamiento real del consumidor.



## Metodolog√≠a

###  Conjunto de datos utilizado:

Se trabaj√≥ con dos tablas principales:

* **amazon_product.csv**: contiene informaci√≥n de producto (precio original, precio con descuento, porcentaje de descuento y categor√≠a).
* **amazon_review.csv**: incluye calificaciones (rating), n√∫mero de votos y comentarios de usuarios.

###  Procesamiento realizado:

1. **Importaci√≥n y limpieza de datos:**

   * Conversi√≥n de precios y descuentos a formato num√©rico.
   * Eliminaci√≥n de duplicados (**product_id**, **review_id**).
   * Imputaci√≥n de valores nulos (links, rating\_count).
   * Unificaci√≥n de categor√≠as (**category_slim**).

2. **Control de calidad:**

   * Se verific√≥ que los precios fueran positivos.
   * Se detectaron outliers extremos en **actual_price** y se reemplazaron usando t√©cnica de percentiles (Q3 + 1.5 IQR).
   * Se confirmaron tipos de datos consistentes (**float64** para precios y ratings, **int64** para recuentos).

3. **Fusi√≥n de tablas:**

   * Se us√≥ **product_id** como clave com√∫n, logrando una tabla final (**df_amazon_full**) con 1,194 productos evaluados.

4. **An√°lisis exploratorio y segmentaci√≥n:**

   * Se calcularon medidas de tendencia, dispersi√≥n y correlaci√≥n.
   * Se segment√≥ el descuento en 5 niveles para observar patrones.

* **Rating promedio:** 4.08 (de 5)
