```python
# Guarda el original por seguridad
df_amazon_product["category_raw"] = df_amazon_product["category"].astype(str)

# Split por jerarquía
parts = df_amazon_product["category_raw"].str.split("|")

# Nivel superior (top) y hoja (leaf)
df_amazon_product["category_top"]  = parts.str[0]
df_amazon_product["category_leaf"] = parts.str[-1]

print("Ejemplo:")
print(df_amazon_product[["category_raw","category_top","category_leaf"]].head(5))
```

```python
import re

def clean_token(s: str) -> str:
    s = str(s)
    s = s.replace("&", " & ")  # arregla cosas como "Mobiles&Accessories"
    s = s.replace(",", ", ")   # comas con espacio
    s = re.sub(r"(?<=[a-z])(?=[A-Z])", " ", s)  # ComputerAccessories → Computer Accessories
    s = re.sub(r"\s+", " ", s).strip()
    return s.title()  # pone cada palabra con mayúscula

df_amazon_product["category_top_clean"]  = df_amazon_product["category_top"].map(clean_token)
df_amazon_product["category_leaf_clean"] = df_amazon_product["category_leaf"].map(clean_token)
```

```python
top_map = {
    "Computers & Accessories": "Computers & Accessories",
    "Electronics": "Electronics",
    "Home & Kitchen": "Home & Kitchen",
    "Office Products": "Office Products",
    "Musical Instruments": "Musical Instruments",
    "Home Improvement": "Home Improvement",
    "Toys & Games": "Toys & Games",
    "Health & Personal Care": "Health & Personal Care",
    "Car & Motorbike": "Car & Motorbike",
}

df_amazon_product["category_top_clean"] = df_amazon_product["category_top_clean"].apply(
    lambda x: top_map.get(x, x)  # deja igual si no está mapeado
)
```

```python
import re

def clean_price(text):
    # Elimina todo lo que no sea número o punto
    clean_text = re.sub(r"[^\d.]", "", str(text))
    return float(clean_text) if clean_text else None

# Aplicar la limpieza a ambas columnas
df_amazon_product["actual_price"] = df_amazon_product_raw["actual_price"].apply(clean_price)
df_amazon_product["discounted_price"] = df_amazon_product_raw["discounted_price"].apply(clean_price)

# Verificamos
print(df_amazon_product[["actual_price", "discounted_price"]].describe())
```

```python
def resumen_dispersión(variable, nombre):
    print(f"Medidas de dispersión para {nombre}")
    print("Desviación estándar:", variable.std())
    print("Varianza:", variable.var())
    print("Rango intercuartílico (IQR):", variable.quantile(0.75) - variable.quantile(0.25))
    print("-" * 40)

# Aplicar a descuento
resumen_dispersión(df_amazon_full["discount_percentage"], "Descuento (%)")

# Aplicar a rating
resumen_dispersión(df_amazon_full["rating"], "Rating")
```

## Introducción

En el competitivo entorno del comercio electrónico, las estrategias de precios juegan un rol clave en la percepción y decisión de compra de los consumidores. Una táctica común es aplicar descuentos visibles en los productos, lo cual puede influir tanto en el volumen de ventas como en las calificaciones que reciben los productos.

Este proyecto tiene como objetivo responder a una pregunta de negocio concreta:

 **¿Cuánto mayor es el descuento, mejor es el rating del producto en Amazon?**

A partir de esta pregunta, se busca analizar si existe una relación significativa entre el porcentaje de descuento aplicado a un producto y la calificación promedio (rating) otorgada por los usuarios. La respuesta puede ayudar a mejorar las estrategias comerciales, equilibrando descuentos agresivos con la reputación del producto y el comportamiento real del consumidor.



## Metodología

###  Conjunto de datos utilizado:

Se trabajó con dos tablas principales:

* **amazon_product.csv**: contiene información de producto (precio original, precio con descuento, porcentaje de descuento y categoría).
* **amazon_review.csv**: incluye calificaciones (rating), número de votos y comentarios de usuarios.

###  Procesamiento realizado:

1. **Importación y limpieza de datos:**

   * Conversión de precios y descuentos a formato numérico.
   * Eliminación de duplicados (**product_id**, **review_id**).
   * Imputación de valores nulos (links, rating\_count).
   * Unificación de categorías (**category_slim**).

2. **Control de calidad:**

   * Se verificó que los precios fueran positivos.
   * Se detectaron outliers extremos en **actual_price** y se reemplazaron usando técnica de percentiles (Q3 + 1.5 IQR).
   * Se confirmaron tipos de datos consistentes (**float64** para precios y ratings, **int64** para recuentos).

3. **Fusión de tablas:**

   * Se usó **product_id** como clave común, logrando una tabla final (**df_amazon_full**) con 1,194 productos evaluados.

4. **Análisis exploratorio y segmentación:**

   * Se calcularon medidas de tendencia, dispersión y correlación.
   * Se segmentó el descuento en 5 niveles para observar patrones.


##  Resultados

### 1. **Distribución general**

* **Descuento promedio:** 46.5%
* **Rating promedio:** 4.08 (de 5)
* **Moda del rating:** 4.1
* **Cantidad promedio de votos por producto:** 14,295

| Variables relacionadas     | Correlación   |

Conclusiones

No se encontró una relación positiva entre el porcentaje de descuento y el nivel de calificación del producto.
Por el contrario, los productos con descuentos bajos (0-20%) tienden a recibir mejores valoraciones promedio.
Los productos con descuentos extremos (más del 80%) presentan ratings menores y mayor variabilidad (presencia de outliers).
El análisis sugiere que un alto descuento no garantiza una mejor percepción del producto, y podría asociarse a productos con calidad percibida inferior, necesidad de liquidación o estrategias de marketing agresivas.
Recomendaciones

Para empresas como Super Caja o marcas en Amazon:

1. Evitar descuentos extremos en productos bien posicionados: Estos no mejoran la percepción y podrían incluso deteriorarla.

2. Usar descuentos moderados (20-40%) como estrategia de equilibrio: Atraen al cliente sin dañar la reputación del producto.

3. Segmentar campañas promocionales: Analizar por categoría y rating histórico si un producto necesita o no descuento.

4. Priorizar productos con alta calificación para campañas de confianza o branding. No todos los productos se benefician de descuentos agresivos.
